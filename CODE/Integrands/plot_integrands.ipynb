```python
# generate 2 2d grids for the x & y bounds
y, x = np.meshgrid(np.linspace(-1.5+1/(a[2] + a[1]), 2.5+1/(a[2] + a[1]), 100), np.linspace(0.0, 2+1/(a[2] + a[1]), 100))

z = Int1_2
# x and y are bounds, so z should be the value *inside* those bounds.
# Therefore, remove the last value from the z array.
z = z[:-1, :-1]
z_min, z_max = -np.abs(z).max(), np.abs(z).max()

fig, ax = plt.subplots()

c = ax.pcolormesh(x, y, z, cmap='RdYlGn', vmin=z_min, vmax=z_max)
ax.set_title(r'Integrand with $p=1$ and $q=2$ on the $AB$ quadrant')
ax.set_xlabel(r'$\rho$')
ax.set_ylabel(r'$z$')
# set the limits of the plot to the limits of the data
ax.axis([x.min(), x.max(), y.min(), y.max()])
fig.colorbar(c, ax=ax)
plt.savefig('../outputs/pics/int1_2.png')

plt.show()
```

```python
n = np.loadtxt('Density.txt')
dn_dX = np.loadtxt('Density_derivative.txt')
dn_dz = np.loadtxt('Density_dz.txt')
```
